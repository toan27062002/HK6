```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(X, y, learning_rate, num_iterations):
    # Khởi tạo tham số w và b
    w = 0
    b = 0
    m = len(X)  # Số lượng mẫu

    # Lưu giá trị tham số w và b theo từng vòng lặp
    w_history = []
    b_history = []

    # Lưu giá trị hàm mất mát theo từng vòng lặp
    cost_history = []

    for iteration in range(num_iterations):
        # Tính đạo hàm riêng của hàm mất mát theo w và b
        dw = (1/m) * np.sum((w*X + b - y) * X)
        db = (1/m) * np.sum(w*X + b - y)

        # Cập nhật tham số w và b
        w = w - learning_rate * dw
        b = b - learning_rate * db

        # Lưu giá trị tham số w và b
        w_history.append(w)
        b_history.append(b)

        # Tính giá trị hàm mất mát và lưu vào history
        cost = (1/(2*m)) * np.sum((w*X + b - y)**2)
        cost_history.append(cost)

    return w, b, w_history, b_history, cost_history
```

```python
# Tạo dữ liệu mẫu
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = 3*X + np.random.normal(0, 1, 100)

# Biểu diễn dữ liệu
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Biểu đồ dữ liệu')
plt.show()
```

```python
# Tham số học và số lần lặp
learning_rate = 0.01
num_iterations = 1000

# Áp dụng Gradient Descent
w_optimal, b_optimal, w_history, b_history, cost_history = gradient_descent(X, y, learning_rate, num_iterations)
```

```python
# Vẽ biểu đồ hàm mất mát theo từng vòng lặp
plt.plot(range(num_iterations), cost_history)
plt.xlabel('Vòng lặp')
plt.ylabel('Hàm mất mát')
plt.title('Biểu đồ hàm mất mát theo từng vòng lặp')
plt.show()

# Vẽ biểu đồ tham số w và b theo từng vòng lặp
plt.plot(range(num_iterations), w_history, label='w')
plt.plot(range(num_iterations), b_history, label='b')
plt.xlabel('Vòng lặp')
plt.ylabel('Giá trị tham số')
plt.title('Biểu đồ tham số w và b theo từng vòng lặp')
plt.legend()
plt.show()

# In kết quả
print("Tham số tối ưu:")
print("w =", w_optimal)
print("b =", b_optimal)

# Vẽ đường thẳng hồi quy
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Biểu đồ hàm tuyến tính')
plt.plot(X, w_optimal*X + b_optimal, color='red', label='Regression Line')
plt.legend()
plt.show()
```
